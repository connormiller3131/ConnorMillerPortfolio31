## 🛷 Sliding Grading Scale — Composite of MAE, RMSE, and MAPE

**Goal:** Reward teams that are consistently strong across **all three** error metrics on the **2025** holdout data.

### 1) Metrics (lower is better)
Let $y_t$ be actuals and $\hat{y}_t$ predictions for all 2025 days $t \in \mathcal{T}$.

- $\displaystyle \text{MAE}=\frac{1}{|\mathcal{T}|}\sum_{t\in\mathcal{T}}|y_t-\hat{y}_t|$
- $\displaystyle \text{RMSE}=\sqrt{\frac{1}{|\mathcal{T}|}\sum_{t\in\mathcal{T}}(y_t-\hat{y}_t)^2}$
- $\displaystyle \text{MAPE}=\frac{100\%}{|\mathcal{T}_+|}\sum_{t\in\mathcal{T}_+}\left|\frac{y_t-\hat{y}_t}{y_t}\right|$

---

### 2) Ranking procedure
1. **Rank by each metric** separately (MAE, RMSE, MAPE). Lower value = better rank (1 = best).
2. Compute the **average rank**:
   $
   \text{AvgRank}=\frac{\text{rank}_{\text{MAE}}+\text{rank}_{\text{RMSE}}+\text{rank}_{\text{MAPE}}}{3}
   $
3. **Composite ranking**: sort teams by **AvgRank** (ascending).
   - **Tie-breakers (in order):** better MAE rank → better RMSE rank → better MAPE rank → clearer methodology & reproducibility.

---

### 3) Starting score from composite rank
- **Top composite rank:** **100**
- **Lowest composite rank:** **80**
- Teams in between are **linearly spaced** between 100 and 80 by **composite rank order**.

For **N teams** and **composite rank \(r\)** (1 = best):
$
\text{Start}(r)=100-20\cdot\frac{(r-1)}{(N-1)}
$
- With **3 teams** → starting scores are **100, 90, 80** for ranks 1, 2, 3.
- With **5 teams** → **100, 95, 90, 85, 80**.

**Final Grade = Starting Score − (Rubric Deductions).**

---

### 🔒 What You Can and Can’t Use Each Day
A reminder on data leakage!

When predicting **energy production on a given day (t)**:

- ✅ You **can use the weather for that same day (t)** — because weather is known or forecast in advance.  
  Example: If it’s **April 1, 2023**, you can use **April , 2023 weather** (like temperature, wind, sunlight).

- ✅ You **can use past energy and past weather** (from previous days).  
  Example: For **April 1, 2023**, you can use **March 31 energy** and **March 31 weather** (or earlier).  
  You can also use lagged or rolling features like “average temperature over the past 7 days.”

- ❌ You **cannot use today’s energy** (the thing you’re trying to predict) or any future data.  
  So no `TOTAL_t` or `TOTAL_t+1` in your predictors — that would be data leakage. You cannot use the individual energy generation for that same day, either.

---

**In short:**  
> “When predicting today’s energy, you can use today’s weather and anything from the past — but never today’s energy or anything from the future.”

Make sure your feature engineering follows that rule for both the **training data** (before Jan 1 2025) and the **test data** (all 2025 data).

---

**If you want a formula...:**


$
\widehat{E_t} = f(W_t,\, W_{t-1},\, W_{t-2},\, \ldots,\, E_{t-1},\, E_{t-2},\, \ldots)
$

Where:  
- $\widehat{E_t} $ = predicted energy on day *t*  
- $ W_t $ = today’s weather (known or forecast)  
- $ E_{t-1}, E_{t-2}, \ldots $ = past energy production  
- ❌ Do **not** use $ E_t $ or any $ W_{t+1+} $ values (future data)

## Lagged features for modeling

To check your understanding of what lags are, look up! Try to match the values.

COAL = 4.65	... it's like the 4th row... which is the first row with good data quality once all the lags=3 are taken care of!

```python
def create_lag_features(df, lags=3):
    df_lagged = df.copy()
    for col in df.columns:
        for i in range(1, lags + 1):
            df_lagged[f'{col}_lag_{i}'] = df_lagged[col].shift(i)
    return df_lagged

X_train_lagged = create_lag_features(X_train)
X_test_lagged = create_lag_features(X_test)

# Drop rows with NaN values introduced by lagging
X_train_lagged = X_train_lagged.dropna()
y_train_lagged = y_train.loc[X_train_lagged.index]

X_test_lagged = X_test_lagged.dropna()
y_test_lagged = y_test.loc[X_test_lagged.index]


print("Shape of X_train_lagged:", X_train_lagged.shape)
print("Shape of X_test_lagged:", X_test_lagged.shape)

display(X_train_lagged.head())
display(X_test_lagged.head())
```

# Modeling

```python
combined_df = combined_df_sorted.copy()
```

```python
combined_df['TOTAL'] = combined_df['TOTAL'].clip(upper=combined_df['TOTAL'].quantile(0.995))
```

```python
combined_df['srad_mj'] = combined_df['srad_mj'].fillna(0)
combined_df['max_wind_gust_kts'] = combined_df['max_wind_gust_kts'].fillna(0)
```

```python
combined_df['day_of_week'] = combined_df.index.dayofweek
combined_df['is_weekend'] = (combined_df['day_of_week'] >= 5).astype(int)
combined_df['month'] = combined_df.index.month
```

```python
combined_df['month'] = combined_df.index.month
```

```python
combined_df['avg_temp'] = (combined_df['max_temp_f'] + combined_df['min_temp_f']) / 2
combined_df['avg_temp_sqr'] = combined_df['avg_temp'] ** 2
```

```python
combined_df['lag_1'] = combined_df['TOTAL'].shift(1)
combined_df['lag_7'] = combined_df['TOTAL'].shift(7)
combined_df['lag_diff_7'] = combined_df['TOTAL'].shift(1) - combined_df['TOTAL'].shift(7)
```

```python
combined_df['roll_7'] = combined_df['TOTAL'].rolling(7).mean()
```

```python
combined_df = combined_df.dropna()
```

```python
train = combined_df[combined_df.index < "2025-01-01"]
test = combined_df[combined_df.index >= "2025-01-01"]
```

```python
features = [
     'is_weekend', 'month',
    'avg_feel', 'avg_temp_sqr', # Corrected from avg_feel_sqr to avg_temp_sqr
    'lag_1', 'lag_7',
    'roll_7', 'lag_diff_7', 'srad_mj', 'max_wind_gust_kts'
]

X_train = train[features]
y_train = train['TOTAL']

X_test = test[features]
y_test = test['TOTAL']
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(
    n_estimators=400,
    max_depth=9,
    min_samples_split=8,
    min_samples_leaf=4,
    max_features='sqrt',
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)
```

```python
pred_train = model.predict(X_train)
pred_test = model.predict(X_test)
```

I chose the lag features because energy demand follows strong short-term patterns. Yesterday’s usage often influences today, and the weekly cycle repeats itself, which makes the 7-day lag especially useful. Longer lags like 14 and 30 days help the model pick up on broader monthly or seasonal rhythms that aren’t visible day-to-day. I also included rolling averages and rolling min/max values to smooth out noise in the data and show the model the recent trend instead of reacting to a single unusual day. The 7-day rolling window captures weekly behavior, while the 30-day window gives a sense of seasonal momentum.

For the Random Forest hyperparameters, I used a moderate number of trees to keep the model stable without making it unnecessarily complex. I limited the tree depth and increased the minimum split and leaf sizes so the trees wouldn’t memorize the training data, which helps the model generalize better to unseen days. I kept the standard “sqrt” setting for feature sampling because it gives each tree a slightly different perspective, improving overall stability.

I chose a Random Forest instead of a neural network because it performs very well on structured weather-and-time data and is easier to interpret. For tuning the Random Forest, I adjusted the number of trees, tree depth, and minimum split/leaf sizes to balance flexibility and generalization. Moderate depth and slightly larger leaf sizes helped prevent overfitting, while increasing the number of trees improved stability. I tested different combinations of features and kept the ones that consistently strengthened test performance. Overall, the model and tuning decisions were designed to capture the key weather and temporal patterns driving energy use while producing reliable, interpretable forecasts.

To handle missing data, I focused on keeping the dataset consistent while avoiding unnecessary drops in observations. The weather variables srad_mj and max_wind_gust_kts contained occasional missing values, but these gaps were small and appeared randomly throughout the dataset rather than in long blocks. Because these features are not the primary drivers of energy demand, I used simple imputation by filling missing values with zero. This approach prevents the model from losing entire days of data while still reflecting that no solar radiation or no recorded gust may have occurred on those days. For the lag and rolling features, missing values naturally appear at the beginning of the dataset, so I filtered out those initial rows after engineering features. This avoids giving the model incomplete or misleading information without affecting the integrity of the dataset. Overall, the combination of light imputation and targeted filtering keeps the dataset complete, preserves meaningful patterns, and prevents missing values from weakening model performance.

# Conclusion

I discovered strong seasonal and temperature-driven patterns in energy use. Demand consistently rose during extreme heat and cold, while mild temperatures produced much lower and more stable usage. This made weather-based features essential for capturing the shape of demand across the year

The hardest part to model was the sudden spikes and drops in energy production that didn’t follow the usual seasonal or weekly patterns. These jumps often happened on random days where weather alone couldn’t explain what was going on. Because I didn’t have additional information like grid outages, industrial activity, or equipment failures, the model had no way to “learn” why those points were different. As a result, the predictions tended to under or over shoot on those unusual days, even when the rest of the forecast looked strong

Feature engineering ended up having one of the biggest impacts on how well the model performed. Adding simple things like lags, rolling averages, and temperature-based features helped the model see the structure that wasn’t obvious from raw data. Some features, like lag-1 and rolling-7, were far more influential than I expected, and removing them noticeably hurt performance. Overall, engineering the right features mattered more than changing the model itself, because it gave the algorithm the information it actually needed to recognize patterns.

I also learned that some forms of energy production don’t respond much to weather, which impacts predictive accuracy. Sources like nuclear or refuse stay almost constant, while gas and total usage swing dramatically. This mismatch made it harder for the model to generalize in periods with unusual conditions.

If I had more data, I would focus on adding more real-world context rather than making the model more complicated. Extra information like economic activity, population behavior, or specific industrial events would help explain the spikes that weather alone can’t capture. With more time, I would also test different lag lengths and rolling windows to see which ones best fit the patterns in the data.
